```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

si = pd.read_csv('si.csv')

si.head()
```

```python
si.dtypes
```

```python
#EXCLUINDO AS COLUNAS [PAI_FALECIDO, MAE_FALECIDO]
si.drop(si.columns[[1, 2]],axis=1, inplace=True)

si.head()
```

```python
# Usando somentes as tuplas em que o idioma é inglês ou NULL

si = si.loc[(si['DESC_IDIOMA']=='Inglês') | (si['DESC_IDIOMA'].isnull())]

#si
```

```python
#RENOMEANDO CAMPOS DAS 4 HABILIDADES EM INGLÊS
si = si.rename(columns = {'NIVEL_FALA':'INGLES_FALA'})
si = si.rename(columns = {'NIVEL_COMPREENSAO':'INGLES_COMPREENSAO'})
si = si.rename(columns = {'NIVEL_ESCRITA':'INGLES_ESCRITA'})
si = si.rename(columns = {'NIVEL_LEITURA':'INGLES_LEITURA'})

#EXCLUINDO O ATRIBUTO DESC_IDIOMA
si.drop('DESC_IDIOMA',axis=1, inplace=True)

si
```

```python
# Atribuindo 0 para o nivel de idioma que está como NULL

si.loc[si['INGLES_FALA'].isnull(), 'INGLES_FALA'] = 0
si.loc[si['INGLES_COMPREENSAO'].isnull(), 'INGLES_COMPREENSAO'] = 0
si.loc[si['INGLES_ESCRITA'].isnull(), 'INGLES_ESCRITA'] = 0
si.loc[si['INGLES_LEITURA'].isnull(), 'INGLES_LEITURA'] = 0

si
```

```python
# Verificando valores existentes na coluna de situação da matricula

si['SIT_MATRICULA'].unique()
```

```python
# Normalizando a coluna de situação da matricula, sendo:
# 0 = matriculado
# 1 = evasão (compreende a categoria 2, 9 e 10)

si.loc[si['SIT_MATRICULA']==2,'SIT_MATRICULA'] = 1
si.loc[si['SIT_MATRICULA']==9,'SIT_MATRICULA'] = 1
si.loc[si['SIT_MATRICULA']==10,'SIT_MATRICULA'] = 1

si['SIT_MATRICULA'].unique()
```

```python
#ATRIBUINDO A MÉDIA ARRENDONDADA AOS MISSING VALUES NA RENDA FAMILIAR
# Considerando que 0 = aluno optou por não informar a renda
# No calculo foi desconsiderado quem está com renda familiar = 0

media_ajustada = round(si.loc[si['RENDA_FAMILIAR']!=0,'RENDA_FAMILIAR'].mean())

si.loc[si['RENDA_FAMILIAR'].isnull(),'RENDA_FAMILIAR']=media_ajustada

si
```

```python
# Verificando valores existentes na coluna de tipo escola origem

si['TIPO_ESCOLA_ORIGEM'].unique()
```

```python
# Atribuindo uma letra para as tuplas com valor vazio, para identificar que o valor era null
# N é a letra definida para identificar que o valor era null

si.loc[si['TIPO_ESCOLA_ORIGEM'].isnull(),'TIPO_ESCOLA_ORIGEM']='N'

si['TIPO_ESCOLA_ORIGEM'].unique()
```

```python
#DISCRETIZAÇÃO DO TIPO_ESCOLA_ORIGEM
si.loc[si['TIPO_ESCOLA_ORIGEM']=='P','TIPO_ESCOLA_ORIGEM'] = '0'
si.loc[si['TIPO_ESCOLA_ORIGEM']=='E','TIPO_ESCOLA_ORIGEM'] = '1'
si.loc[si['TIPO_ESCOLA_ORIGEM']=='F','TIPO_ESCOLA_ORIGEM'] = '2'
si.loc[si['TIPO_ESCOLA_ORIGEM']=='N','TIPO_ESCOLA_ORIGEM'] = '3'

si['TIPO_ESCOLA_ORIGEM'] = si['TIPO_ESCOLA_ORIGEM'].astype('float64')

si['TIPO_ESCOLA_ORIGEM'].hist()
```

```python
#ATRIBUINDO A MÉDIA ARRENDONDADA AOS MISSING VALUES NA RENDA_PER_CAPITA

media = round(si.loc[si['RENDA_PER_CAPITA']!=10,'RENDA_PER_CAPITA'].mean())

si.loc[si['RENDA_PER_CAPITA'].isnull(),'RENDA_PER_CAPITA']=media
#media
si['RENDA_PER_CAPITA'].unique()
```

```python
#si.loc[si['COD_ESTADO_CIVIL']==0].count()
si['COD_ESTADO_CIVIL'].unique()
```

```python
#IMPUTANDO O VALOR DE 'S' (SOLTEIRO) AOS MISSING VALUES EM COD_ESTADO_CIVIL

si.loc[si['COD_ESTADO_CIVIL'].isnull(), 'COD_ESTADO_CIVIL']='S'

si['COD_ESTADO_CIVIL'].unique()
```

```python
# Discretização do COD_ESTADO_CIVIL

si.loc[si['COD_ESTADO_CIVIL']=='S','COD_ESTADO_CIVIL']=0
si.loc[si['COD_ESTADO_CIVIL']=='C','COD_ESTADO_CIVIL']=1
si.loc[si['COD_ESTADO_CIVIL']=='D','COD_ESTADO_CIVIL']=2

si['COD_ESTADO_CIVIL'] = si['COD_ESTADO_CIVIL'].astype('float64')
#si.dtypes
si['COD_ESTADO_CIVIL'].unique()
```

```python
# Atribuindo 0 para os valores null do N_FILHOS

si.loc[si['N_FILHOS'].isnull(),'N_FILHOS']=0

# Se existir algum aluno com mais de um filho, atribuir 1
# Discretizando para:
# 0 = Não tem filho
# 1 = Tem filho

si.loc[si['N_FILHOS']>1,'N_FILHOS']=1

si['N_FILHOS'].unique()
```

```python
# Discretizando SEXO
# 0 = Homem (Masculino)
# 1 = Mulher (Feminino)

si.loc[si['SEXO']=='M','SEXO'] = 0
si.loc[si['SEXO']=='F','SEXO'] = 1

si['SEXO'] = si['SEXO'].astype('float64')

#si['SEXO'].unique()
si['SEXO'].hist()
```

```python
si['PROFISSAO'].unique()
```

```python
# Discretizando PROFISSAO
# 0 = null e estudantes
# 1 = todo o resto

si.loc[si['PROFISSAO'].isnull(),'PROFISSAO']=0
si.loc[si['PROFISSAO']=='ESTUDANTE','PROFISSAO']=0
si.loc[si['PROFISSAO']=='Estudante','PROFISSAO']=0
si.loc[si['PROFISSAO']!=0,'PROFISSAO']=1

si['PROFISSAO'] = si['PROFISSAO'].astype('float64')

si['PROFISSAO'].unique()
```

```python
# Atribuindo 0 para os valores null em NOTA

si.loc[si['NOTAS'].isnull(),'NOTAS'] = 0.0
```

```python
si['NOTAS'].unique()
```

```python
#RETIRANDO AS NOTAS QUE CONTÉM UMA DATA (OUTLEIRS)
#SELECIONANDO TODOS OS CAMPOS QUE CONTÉM '/' E ATRIBUINDO A NEGAÇÃO (~) AO DATASET
si = si[~si['NOTAS'].str.contains("/",na=False)]

si['NOTAS'].unique()
```

```python
#AJUSTANDO TIPOS DA NOTA. CONVERTENDO DE OBJECT PARA FLOAT

si['NOTAS'] = pd.to_numeric(si['NOTAS'])

#CRIANDO CAMPO SITUAÇÃO E ATRIBUINDO O VALOR ZERO PARA TODOS

si['APROVADO'] = 0.0

si['NOTAS'].unique()
```

```python
#ATRIBUINDO APROVADO (1) AO CAMPO SITUAÇÃO COM BASE NA NOTA

si.loc[si['NOTAS']>=6.0,'APROVADO']=1.0

si['APROVADO'].unique()
```

```python
# Atribuindo zero para os null em N_FALTAS

si.loc[si['N_FALTAS'].isnull(), 'N_FALTAS']=0

si['N_FALTAS'].unique()
```

```python
# Calculando a porcentagem de faltas
# Primeiro transforma a carga horaria em minutos(*60), depois divide por quantos minutos equivale uma aula(/45), 
# Chegando no resultado da quantidade de aulas
si['PORCENTAGEM_FALTAS'] = (si['N_FALTAS'] / (si['CARGA_HOR'] * 60 / 45)) * 100

# Atribuindo como reprovado o aluno que excedeu a porcentagem de faltas

si.loc[si['PORCENTAGEM_FALTAS']>75.0,'APROVADO']=0

#si['PORCENTAGEM_FALTAS'].unique()

# Na turma de SI, ninguem dos aprovados por nota, foi reprovado por falta
#si.loc[(si['PORCENTAGEM_FALTAS']>75.0) & (si['APROVADO']==1)]

# Removendo o atributo criado para auxiliar
si.drop('PORCENTAGEM_FALTAS', axis=1, inplace=True)

#si
```

```python
# Existe apenas alunos com o estado de Goiás
si['DESC_ESTADO'].unique()

# Removendo estado, pois existe apenas um estado
si.drop('DESC_ESTADO', axis=1, inplace=True)
```

```python
si['DESC_CIDADE'].unique()
```

```python
# Tratando cidades
# C = Capital(Goiânia)
# G = Grande Goiânia(Aparecida de Goiânia, Senador Canedo e Trindade)
# I = Interior de Goiás(Outras cidades de Goiás)

si.loc[si['DESC_CIDADE']=='Goiânia', 'DESC_CIDADE'] = 'C'
si.loc[si['DESC_CIDADE']=='Aparecida de Goiânia', 'DESC_CIDADE'] = 'G'
si.loc[si['DESC_CIDADE']=='Senador Canedo', 'DESC_CIDADE'] = 'G'
si.loc[si['DESC_CIDADE']=='Anicuns', 'DESC_CIDADE'] = 'I'

si['DESC_CIDADE'].unique()
```

```python
# Discretizando cidades
# 0 = Capital
# 1 = Grande Goiânia
# 2 = Interior de Goiás

si.loc[si['DESC_CIDADE']=='C', 'DESC_CIDADE'] = 0
si.loc[si['DESC_CIDADE']=='G', 'DESC_CIDADE'] = 1
si.loc[si['DESC_CIDADE']=='I', 'DESC_CIDADE'] = 2

si['DESC_CIDADE'] = si['DESC_CIDADE'].astype('float64')

si['DESC_CIDADE'].hist()
```

```python
si.dtypes
```

```python
si['IDADE_INGRESSO'] = si['ANO_INGRESSO'].astype('float64') - si['DT_NASCIMENTO'].str[:4].astype('float64')

si['IDADE_INGRESSO'].unique()
```

```python
#VERIFICANDO ASSOCIAÇÃO DO ANO_INGRESSO COM O ANO_CONCLUSAO_2_GRAU
grau2 = si[['IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU','APROVADO']]

sb.pairplot(grau2.dropna(), hue='APROVADO')
```

```python
#EXCLUINDO OUTLIER DA SÉRIE HISTÓRICA E REGERANDO PLOT

grau2_clean = grau2.loc[si['IDADE_INGRESSO']<50]
#sb.pairplot(grau2_clean.dropna(), hue='APROVADO')

sb.lmplot(x="IDADE_INGRESSO", 
          y="ANO_CONCLUSAO_2_GRAU", 
          col="APROVADO", 
          hue="APROVADO", 
          data=grau2_clean, 
          col_wrap=2, ci=None, 
          palette="muted", 
          size=4, scatter_kws={"s": 50, "alpha": 1})
```

```python
# Regressão Linear

import numpy as np

from sklearn import linear_model

grau3 = si[['IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU']]
grau3 = grau3.loc[si['IDADE_INGRESSO']<40]

#CRANDO BASE DE TREINO SEM ANO_CONCLUSAO_2_GRAU NULOS 
treino_pd = grau3.loc[~grau3['ANO_CONCLUSAO_2_GRAU'].isnull()]

#CRANDO BASE DE REGRESSÃO COM ANO_CONCLUSAO_2_GRAU NULOS (PANDAS pd)
prdict_pd = grau3.loc[grau3['ANO_CONCLUSAO_2_GRAU'].isnull()]

#CRIANDO NUMPY-ARRAYS
#treino_np = treino_pd.as_matrix(columns={'IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU'})
#prdict_np = prdict_pd.as_matrix(columns={'IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU'})

treino_np = treino_pd.as_matrix()
prdict_np = prdict_pd.as_matrix()

prdict_np
```

```python
treino_np
```

```python
#CRIANDO UM OBJETO DE REGRESSÃO LINEAR DO SCIKIT-LEARN
regr = linear_model.LinearRegression()

#SEPARANDO O X E O Y PARA A REGRESSÃO DO CONJUNTO DE TREINO
idade_ingresso_treino = treino_np[:, np.newaxis, 0]
ano_concussao_treino = treino_np[:, np.newaxis, 1]

#TREINANDO COM O CONJUNTO DE TREINO X E Y PARA REGRESSÃO
regr.fit(idade_ingresso_treino, ano_concussao_treino)
```

```python
#IDADES_INGRESSO QUE NÃO TEM ANO_CONCLUSAO_2_GRAU PARA A REGRESSAO
#SELECIONANDO A IDADE_INGRESSO DE TESTE PARA REALIZAR PREDIÇÃO DO ANO_CONCLUSAO_2_GRAU

idade_ingresso_predict = prdict_np[:, np.newaxis, 0]
idade_ingresso_predict
```

```python
#REALIZANDO A REGRESSÃO

ano_predict = regr.predict(idade_ingresso_predict)

ano_predict
```

```python
#JUNTANDO A REGRESSÃO DO ANO_CONCLUSAO_2_GRAU COM A IDADE_INGRESSO

prdict_np[:, np.newaxis, 1] = ano_predict

prdict_np
```

```python
# Salvando a predição

si.loc[si['ANO_CONCLUSAO_2_GRAU'].isnull(), 'ANO_CONCLUSAO_2_GRAU'] = prdict_np[:, np.newaxis, 1]

si['ANO_CONCLUSAO_2_GRAU'].unique()
```

```python
si.dtypes
```

```python
# Excluindo colunas: DT_NASCIMENTO, COD_INICIACAO_CIENTIFICA, CARGA_HOR, QTDE_BOLSAS, QTDE_INICIACOES

si.drop('DT_NASCIMENTO',axis=1, inplace=True)
si.drop('COD_INICIACAO_CIENTIFICA',axis=1, inplace=True)
si.drop('CARGA_HOR',axis=1, inplace=True)
si.drop('QTDE_BOLSAS',axis=1, inplace=True)
si.drop('QTDE_INICIACOES',axis=1, inplace=True)

si.dtypes
```

```python
# Salvando

si.to_csv('si_clean.csv', index=False)
```
