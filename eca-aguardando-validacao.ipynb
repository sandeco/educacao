```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

eca = pd.read_csv('eca2.csv')

eca.head()
```

```python
eca.dtypes
```

```python
#EXCLUINDO AS COLUNAS [PAI_FALECIDO, MAE_FALECIDO]
eca.drop(eca.columns[[1, 2]],axis=1, inplace=True)

eca.head()
```

```python
# Usando somentes as tuplas em que o idioma é inglês ou NULL

eca = eca.loc[(eca['DESC_IDIOMA']=='Inglês') | (eca['DESC_IDIOMA'].isnull())]

#eca
```

```python
#RENOMEANDO CAMPOS DAS 4 HABILIDADES EM INGLÊS
eca = eca.rename(columns = {'NIVEL_FALA':'INGLES_FALA'})
eca = eca.rename(columns = {'NIVEL_COMPREENSAO':'INGLES_COMPREENSAO'})
eca = eca.rename(columns = {'NIVEL_ESCRITA':'INGLES_ESCRITA'})
eca = eca.rename(columns = {'NIVEL_LEITURA':'INGLES_LEITURA'})

#EXCLUINDO O ATRIBUTO DESC_IDIOMA
eca.drop('DESC_IDIOMA',axis=1, inplace=True)

eca
```

```python
# Atribuindo 0 para o nivel de idioma que está como NULL

eca.loc[eca['INGLES_FALA'].isnull(), 'INGLES_FALA'] = 0
eca.loc[eca['INGLES_COMPREENSAO'].isnull(), 'INGLES_COMPREENSAO'] = 0
eca.loc[eca['INGLES_ESCRITA'].isnull(), 'INGLES_ESCRITA'] = 0
eca.loc[eca['INGLES_LEITURA'].isnull(), 'INGLES_LEITURA'] = 0

eca
```

```python
# Verificando valores existentes na coluna de situação da matricula

#Verificar com o Sanderson o que fazer com as categorias 13, 14, 18 e 20
eca['SIT_MATRICULA'].unique()
```

```python
eca['RENDA_FAMILIAR'].unique()
```

```python
#ATRIBUINDO A MÉDIA ARRENDONDADA AOS MISSING VALUES NA RENDA FAMILIAR
# Considerando que 0 = aluno optou por não informar a renda
# No calculo foi desconsiderado quem está com renda familiar = 0

media_ajustada = round(eca.loc[eca['RENDA_FAMILIAR']!=0,'RENDA_FAMILIAR'].mean())

eca.loc[eca['RENDA_FAMILIAR'].isnull(),'RENDA_FAMILIAR']=media_ajustada

eca
```

```python
# Verificando valores existentes na coluna de tipo escola origem

eca['TIPO_ESCOLA_ORIGEM'].unique()
```

```python
#AJUSTANDO ERRO DE DIGITAÇÃO NO TIPO_ESCOLA_ORIGEM. DE ZERO PARA A LETRA 'O'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='0','TIPO_ESCOLA_ORIGEM'] = 'O'

eca['TIPO_ESCOLA_ORIGEM'].unique()
```

```python
# Atribuindo uma letra para as tuplas com valor vazio, para identificar que o valor era null
# N é a letra definida para identificar que o valor era null

eca.loc[eca['TIPO_ESCOLA_ORIGEM'].isnull(),'TIPO_ESCOLA_ORIGEM']='N'

eca['TIPO_ESCOLA_ORIGEM'].unique()
```

```python
#DISCRETIZAÇÃO DO TIPO_ESCOLA_ORIGEM
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='P','TIPO_ESCOLA_ORIGEM'] = '0'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='E','TIPO_ESCOLA_ORIGEM'] = '1'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='F','TIPO_ESCOLA_ORIGEM'] = '2'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='O','TIPO_ESCOLA_ORIGEM'] = '3'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='I','TIPO_ESCOLA_ORIGEM'] = '4'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='M','TIPO_ESCOLA_ORIGEM'] = '5'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='C','TIPO_ESCOLA_ORIGEM'] = '6'
eca.loc[eca['TIPO_ESCOLA_ORIGEM']=='N','TIPO_ESCOLA_ORIGEM'] = '7'

eca['TIPO_ESCOLA_ORIGEM'] = eca['TIPO_ESCOLA_ORIGEM'].astype('float64')

eca['TIPO_ESCOLA_ORIGEM'].hist()
```

```python
#ATRIBUINDO A MÉDIA ARRENDONDADA AOS MISecaNG VALUES NA RENDA_PER_CAPITA

media = round(eca.loc[eca['RENDA_PER_CAPITA']!=10,'RENDA_PER_CAPITA'].mean())

eca.loc[eca['RENDA_PER_CAPITA'].isnull(),'RENDA_PER_CAPITA']=media
#media
eca['RENDA_PER_CAPITA'].unique()
```

```python
#eca.loc[si['COD_ESTADO_CIVIL']==0].count()
eca['COD_ESTADO_CIVIL'].unique()
```

```python
#IMPUTANDO O VALOR DE 'S' (SOLTEIRO) AOS MISSING VALUES EM COD_ESTADO_CIVIL

eca.loc[eca['COD_ESTADO_CIVIL'].isnull(), 'COD_ESTADO_CIVIL']='S'

eca['COD_ESTADO_CIVIL'].unique()
```

```python
# Discretização do COD_ESTADO_CIVIL

eca.loc[eca['COD_ESTADO_CIVIL']=='S','COD_ESTADO_CIVIL']=0
eca.loc[eca['COD_ESTADO_CIVIL']=='C','COD_ESTADO_CIVIL']=1
eca.loc[eca['COD_ESTADO_CIVIL']=='UE','COD_ESTADO_CIVIL']=2

eca['COD_ESTADO_CIVIL'] = eca['COD_ESTADO_CIVIL'].astype('float64')
#eca.dtypes
eca['COD_ESTADO_CIVIL'].unique()
```

```python
eca['N_FILHOS'].unique()
```

```python
# Atribuindo 0 para os valores null do N_FILHOS

eca.loc[eca['N_FILHOS'].isnull(),'N_FILHOS']=0

# Se existir algum aluno com mais de um filho, atribuir 1
# Discretizando para:
# 0 = Não tem filho
# 1 = Tem filho

eca.loc[eca['N_FILHOS']>1,'N_FILHOS']=1

eca['N_FILHOS'].unique()
```

```python
# Discretizando SEXO
# 0 = Homem (Masculino)
# 1 = Mulher (Feminino)

eca.loc[eca['SEXO']=='M','SEXO'] = 0
eca.loc[eca['SEXO']=='F','SEXO'] = 1

eca['SEXO'] = eca['SEXO'].astype('float64')

#eca['SEXO'].unique()
eca['SEXO'].hist()
```

```python
eca['PROFISSAO'].unique()
```

```python
# Discretizando PROFISSAO
# 0 = null e estudantes
# 1 = todo o resto

eca.loc[eca['PROFISSAO'].isnull(),'PROFISSAO']=0
eca.loc[eca['PROFISSAO']=='ESTUDANTE','PROFISSAO']=0
eca.loc[eca['PROFISSAO']=='Estudante','PROFISSAO']=0
eca.loc[eca['PROFISSAO']!=0,'PROFISSAO']=1

eca['PROFISSAO'] = eca['PROFISSAO'].astype('float64')

eca['PROFISSAO'].unique()
```

```python
eca['NOTAS'].unique()
```

```python
# Atribuindo 0 para os valores null em NOTA

eca.loc[eca['NOTAS'].isnull(),'NOTAS'] = 0.0
```

```python
#RETIRANDO AS NOTAS QUE CONTÉM UMA DATA (OUTLEIRS)
#SELECIONANDO TODOS OS CAMPOS QUE CONTÉM '/' E ATRIBUINDO A NEGAÇÃO (~) AO DATASET
eca = eca[~eca['NOTAS'].str.contains("/",na=False)]

eca['NOTAS'].unique()
```

```python
#AJUSTANDO TIPOS DA NOTA. CONVERTENDO DE OBJECT PARA FLOAT

eca['NOTAS'] = pd.to_numeric(eca['NOTAS'])

#CRIANDO CAMPO ecaTUAÇÃO E ATRIBUINDO O VALOR ZERO PARA TODOS

eca['APROVADO'] = 0.0

eca['NOTAS'].unique()
```

```python
#ATRIBUINDO APROVADO (1) AO CAMPO ecaTUAÇÃO COM BASE NA NOTA

eca.loc[eca['NOTAS']>=6.0,'APROVADO']=1.0

eca['APROVADO'].unique()
```

```python
eca['N_FALTAS'].unique()
```

```python
# Atribuindo zero para os null em N_FALTAS

eca.loc[eca['N_FALTAS'].isnull(), 'N_FALTAS']=0
```

```python
# Calculando a porcentagem de faltas
# Primeiro transforma a carga horaria em minutos(*60), depois divide por quantos minutos equivale uma aula(/45), 
# Chegando no resultado da quantidade de aulas
eca['PORCENTAGEM_FALTAS'] = (eca['N_FALTAS'] / (eca['CARGA_HOR'] * 60 / 45)) * 100

# Atribuindo como reprovado o aluno que excedeu a porcentagem de faltas

eca.loc[eca['PORCENTAGEM_FALTAS']>75.0,'APROVADO']=0

#eca['PORCENTAGEM_FALTAS'].unique()

# Na turma de eca, ninguem dos aprovados por nota, foi reprovado por falta
#eca.loc[(eca['PORCENTAGEM_FALTAS']>75.0) & (eca['APROVADO']==1)]

# Removendo o atributo criado para auxiliar
eca.drop('PORCENTAGEM_FALTAS', axis=1, inplace=True)

#eca
```

```python
# Existe apenas alunos com o estado de Goiás
eca['DESC_ESTADO'].unique()
```

```python
# Verificando a quantidade de tuplas contendo '--' no DES_ESTADO

eca.loc[eca['DESC_ESTADO']=='--'].count()
```

```python
# Como a quantidade de tuplas contendo '--' é baixa, é possivel excluir elas sem prejudicar a base de dados

eca = eca.loc[eca['DESC_ESTADO']!='--']

eca['DESC_ESTADO'].unique()
```

```python
eca['DESC_CIDADE'].unique()
```

```python
# Tratando cidades
# C = Capital(Goiânia)
# G = Grande Goiânia(Aparecida de Goiânia, Senador Canedo e Trindade)
# I = Interior de Goiás(Outras cidades de Goiás)
# O = Outro estado(Cidades que não são do estado de Goiás)

eca.loc[eca['DESC_CIDADE']=='Goiânia', 'DESC_CIDADE'] = 'C'
eca.loc[eca['DESC_CIDADE']=='Aparecida de Goiânia', 'DESC_CIDADE'] = 'G'
eca.loc[eca['DESC_CIDADE']=='Senador Canedo', 'DESC_CIDADE'] = 'G'
eca.loc[eca['DESC_CIDADE']=='Trindade', 'DESC_CIDADE'] = 'G'
eca.loc[(eca['DESC_ESTADO']=='Goiás') & (eca['DESC_CIDADE']!='C') & (eca['DESC_CIDADE']!='G'), 'DESC_CIDADE'] = 'I'
eca.loc[eca['DESC_ESTADO']!='Goiás', 'DESC_CIDADE']= 'O'

eca['DESC_CIDADE'].unique()
```

```python
# Removendo o atributo estado
eca.drop('DESC_ESTADO', axis=1, inplace=True)
```

```python
# Discretizando cidades
# 0 = Capital
# 1 = Grande Goiânia
# 2 = Interior de Goiás
# 3 = Outro estado

eca.loc[eca['DESC_CIDADE']=='C', 'DESC_CIDADE'] = 0
eca.loc[eca['DESC_CIDADE']=='G', 'DESC_CIDADE'] = 1
eca.loc[eca['DESC_CIDADE']=='I', 'DESC_CIDADE'] = 2
eca.loc[eca['DESC_CIDADE']=='O', 'DESC_CIDADE'] = 3

eca['DESC_CIDADE'] = eca['DESC_CIDADE'].astype('float64')

eca['DESC_CIDADE'].hist()
```

```python
eca.dtypes
```

```python
eca['IDADE_INGRESSO'] = eca['ANO_INGRESSO'].astype('float64') - eca['DT_NASCIMENTO'].str[:4].astype('float64')

eca['IDADE_INGRESSO'].unique()
```

```python
#VERIFICANDO ASSOCIAÇÃO DO ANO_INGRESSO COM O ANO_CONCLUSAO_2_GRAU
grau2 = eca[['IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU','APROVADO']]

sb.pairplot(grau2.dropna(), hue='APROVADO')
```

```python
#EXCLUINDO OUTLIER DA SÉRIE HISTÓRICA E REGERANDO PLOT

grau2_clean = grau2.loc[eca['IDADE_INGRESSO']<50]
#sb.pairplot(grau2_clean.dropna(), hue='APROVADO')

sb.lmplot(x="IDADE_INGRESSO", 
          y="ANO_CONCLUSAO_2_GRAU", 
          col="APROVADO", 
          hue="APROVADO", 
          data=grau2_clean, 
          col_wrap=2, ci=None, 
          palette="muted", 
          size=4, scatter_kws={"s": 50, "alpha": 1})
```

```python
# Regressão Linear

import numpy as np

from sklearn import linear_model

grau3 = eca[['IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU']]
grau3 = grau3.loc[eca['IDADE_INGRESSO']<40]

#CRANDO BASE DE TREINO SEM ANO_CONCLUSAO_2_GRAU NULOS 
treino_pd = grau3.loc[~grau3['ANO_CONCLUSAO_2_GRAU'].isnull()]

#CRANDO BASE DE REGRESSÃO COM ANO_CONCLUSAO_2_GRAU NULOS (PANDAS pd)
prdict_pd = grau3.loc[grau3['ANO_CONCLUSAO_2_GRAU'].isnull()]

#CRIANDO NUMPY-ARRAYS
#treino_np = treino_pd.as_matrix(columns={'IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU'})
#prdict_np = prdict_pd.as_matrix(columns={'IDADE_INGRESSO','ANO_CONCLUSAO_2_GRAU'})

treino_np = treino_pd.as_matrix()
prdict_np = prdict_pd.as_matrix()
treino_np
#eca
```

```python
eca['IDADE_INGRESSO'].unique()
```

```python
#CRIANDO UM OBJETO DE REGRESSÃO LINEAR DO SCIKIT-LEARN
regr = linear_model.LinearRegression()

#SEPARANDO O X E O Y PARA A REGRESSÃO DO CONJUNTO DE TREINO
idade_ingresso_treino = treino_np[:, np.newaxis, 0]
ano_concussao_treino = treino_np[:, np.newaxis, 1]

#TREINANDO COM O CONJUNTO DE TREINO X E Y PARA REGRESSÃO
regr.fit(idade_ingresso_treino, ano_concussao_treino)
```

```python
#IDADES_INGRESSO QUE NÃO TEM ANO_CONCLUSAO_2_GRAU PARA A REGRESSAO
#SELECIONANDO A IDADE_INGRESSO DE TESTE PARA REALIZAR PREDIÇÃO DO ANO_CONCLUSAO_2_GRAU

idade_ingresso_predict = prdict_np[:, np.newaxis, 0]
idade_ingresso_predict
```

```python
#REALIZANDO A REGRESSÃO

ano_predict = regr.predict(idade_ingresso_predict)

ano_predict
```

```python
#JUNTANDO A REGRESSÃO DO ANO_CONCLUSAO_2_GRAU COM A IDADE_INGRESSO

prdict_np[:, np.newaxis, 1] = ano_predict

prdict_np
```

```python
# Salvando a predição

eca.loc[eca['ANO_CONCLUSAO_2_GRAU'].isnull(), 'ANO_CONCLUSAO_2_GRAU'] = prdict_np[:, np.newaxis, 1]

eca['ANO_CONCLUSAO_2_GRAU'].unique()
```
