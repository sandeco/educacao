```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

#CARREGANDO DATASET INFORMANDO FORMATO DOS 'MISSING VALUES'
eca = pd.read_csv('eca.csv', na_values=['NULL'], dtype={"MATRICULA" : "object"})

eca.head()
```
